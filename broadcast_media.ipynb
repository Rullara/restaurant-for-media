# 수요미식회 방문자리뷰수 TOP30 #

```python
from selenium import webdriver
import re

driver = webdriver.Chrome('C:/Temp/chromedriver')
driver.implicitly_wait(3) 
driver.get('https://map.naver.com/v5/search/%EC%88%98%EC%9A%94%EB%AF%B8%EC%8B%9D%ED%9A%8C') 
import time

names_list = []
reviews_list = []
review_counts = []

stopFlag = False
time.sleep(2)
driver.switch_to.frame("searchIframe")

while True :
    count = 9
    while True :
        try :
            driver.execute_script("var su = arguments[0]; var dom=document.querySelectorAll('#_pcmap_list_scroll_container>ul>li')[su]; dom.scrollIntoView();", count)
            time.sleep(2)
        except :        
            break
        count += 10
    names = driver.find_elements_by_css_selector("span._3Yilt")
    reviews = driver.find_elements_by_css_selector("#_pcmap_list_scroll_container > ul > li > div.Ow5Yt > a > div > span._3Yzhl:nth-child(2)")

    for i in names:
        names_list.append(i.text)
    for i in reviews:
        reviews_list.append(i.text)
        
    first_list = driver.find_element_by_css_selector("#_pcmap_list_scroll_container > ul > li:nth-child(1) > div.Ow5Yt > a:nth-child(1) > div._1z7ih > div > span")
    linkurl = '#app-root > div > div._2lx2y > div._2ky45 > a:nth-child(7)'
    linkNum = driver.find_element_by_css_selector(linkurl)  
    linkNum.click() 
    time.sleep(5)
    second_list = driver.find_element_by_css_selector("#_pcmap_list_scroll_container > ul > li:nth-child(1) > div.Ow5Yt > a:nth-child(1) > div._1z7ih > div > span")
    
    if first_list == second_list:
        stropFlag = True
        break
        
names_list.remove('징광문화')   # 방문자리뷰 없는 식당 제외 
```

```python
import pandas as pd

r = re.compile(".*방문자리뷰")
new_reviews_list = list(filter(r.match, reviews_list))                                   # 방문자 리뷰수 정보만 추리기 

review_counts = [re.sub(r"방문자리뷰",'', counts) for counts in new_reviews_list]       # 방문자 리뷰수 숫자값만 추리기 
review_counts = [re.sub(r"\W",'', counts) for counts in review_counts]                  # , 등 숫자가 아닌거 없애기 

for i in range(0, len(review_counts)):
    review_counts[i] = int(review_counts[i])                                             # 리스트에있는 모든 string형 자료를  integer형으로 변환하기 

df = pd.DataFrame(list(zip(names_list, review_counts)), columns = ['식당명', '리뷰수'])    # 식당명과 리뷰수를 Column으로 가지는 데이터프레임 만들기 
pd.set_option("display.max_rows", None, "display.max_columns", None)                       # display 했을때 중간에 생략되지 않고 모든 정보 보이게 하기 

sorted_df = df.sort_values(by=['리뷰수'], ascending=False)                                   # 데이터프레임 descending order로 sorting 하기 

display(sorted_df[:30])                                                                    # 리뷰수 상위 30개의 식당 정보만 출력 

#sorted_df.to_csv('project_data/수요미식회.csv')

#print(len(names_list))
#print(len(review_counts))   
```
